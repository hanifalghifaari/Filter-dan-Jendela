```python
import numpy as np
import scipy.signal as signal
import librosa
import matplotlib.pyplot as plt
import soundfile as sf
import os

# Path file audio input
file_path = r'D:\APAKEK\Indonesia Raya.wav'

# Path folder untuk menyimpan hasil
output_folder = r'D:\APAKEK\hasil'

# Cek apakah folder output ada
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    print(f"Folder dibuat: {output_folder}")

# Cek apakah file audio ada
if not os.path.exists(file_path):
    print(f"File tidak ditemukan: {file_path}")
else:
    # Muat file audio
    y, sr = librosa.load(file_path, sr=None)

    # Desain high-pass FIR filter dengan jendela rectangular
    numtaps = 101  # Jumlah tap filter
    highcut = 1000  # Frekuensi cut-off atas (Hz)
    fir_coeff = signal.firwin(numtaps, highcut, pass_zero=False, window='boxcar', fs=sr)

    # Frequency response dari filter FIR
    w, h = signal.freqz(fir_coeff, worN=8000)

    # Terapkan filter FIR pada sinyal
    filtered_signal_fir = signal.lfilter(fir_coeff, 1.0, y)

    # Path file output
    output_path = os.path.join(output_folder, 'filtered_audio_highpass_fir_rectangular.wav')
    sf.write(output_path, filtered_signal_fir, sr)

    # Plot sinyal asli dan sinyal hasil filter
    plt.figure(figsize=(15, 12))

    # Plot sinyal asli
    plt.subplot(4, 1, 1)
    plt.plot(y)
    plt.title('Sinyal Asli')
    plt.xlabel('Sample')
    plt.ylabel('Amplitude')

    # Plot sinyal hasil filter
    plt.subplot(4, 1, 2)
    plt.plot(filtered_signal_fir)
    plt.title('Sinyal Hasil High-pass FIR Filter dengan Jendela Rectangular')
    plt.xlabel('Sample')
    plt.ylabel('Amplitude')

    # Plot impulse response (jendela rectangular)
    plt.subplot(4, 1, 3)
    plt.plot(fir_coeff)
    plt.title('Impulse Response (Jendela Rectangular)')
    plt.xlabel('Sample')
    plt.ylabel('Amplitude')

    # Plot frequency response
    plt.subplot(4, 1, 4)
    plt.plot(w, np.abs(h))
    plt.title('Frequency Response')
    plt.xlabel('Frequency [radians / sample]')
    plt.ylabel('Amplitude')

    plt.tight_layout()
    plt.show()

    print(f"Sinyal hasil filter disimpan di: {output_path}")
```

```python
import numpy as np
import scipy.signal as signal
import librosa
import matplotlib.pyplot as plt
import soundfile as sf
import os

# Path file audio input
file_path = r'D:\APAKEK\Indonesia Raya.wav'

# Path folder untuk menyimpan hasil
output_folder = r'D:\APAKEK\hasil'

# Cek apakah folder output ada
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    print(f"Folder dibuat: {output_folder}")

# Cek apakah file audio ada
if not os.path.exists(file_path):
    print(f"File tidak ditemukan: {file_path}")
else:
    # Muat file audio
    y, sr = librosa.load(file_path, sr=None)  # sr=None agar menggunakan sr asli dari file

    # Desain high-pass Butterworth filter
    cutoff = 1000  # Frekuensi cut-off (Hz)
    order = 4  # Orde filter Butterworth

    # Frekuensi Nyquist
    nyquist = 0.5 * sr
    
    # Rancang filter Butterworth high-pass
    b, a = signal.butter(order, cutoff/nyquist, btype='high')

    # Terapkan filter Butterworth pada sinyal
    filtered_signal_butter = signal.lfilter(b, a, y)

    # Path file output
    output_path = os.path.join(output_folder, 'filtered_audio_highpass_butterworth.wav')
    sf.write(output_path, filtered_signal_butter, sr)

    # Plot sinyal asli dan sinyal hasil filter
    plt.figure(figsize=(15, 10))

    plt.subplot(4, 1, 1)
    plt.plot(y)
    plt.title('Sinyal Asli')
    plt.xlabel('Sample')
    plt.ylabel('Amplitudo')

    plt.subplot(4, 1, 2)
    plt.plot(filtered_signal_butter)
    plt.title('Sinyal Hasil High-pass Butterworth Filter')
    plt.xlabel('Sample')
    plt.ylabel('Amplitudo')

    # Hitung dan plot respons impuls
    plt.subplot(4, 1, 3)
    impulse_response, _ = signal.impulse((b, a))
    plt.plot(impulse_response)
    plt.title('Impulse Response')
    plt.xlabel('Sample')
    plt.ylabel('Amplitudo')

    # Hitung dan plot respons frekuensi
    plt.subplot(4, 1, 4)
    w, h = signal.freqz(b, a)
    plt.plot(w, np.abs(h))
    plt.title('Frequency Response')
    plt.xlabel('Frequency [radians/sample]')
    plt.ylabel('Amplitudo')

    plt.tight_layout()
    plt.show()

    print(f"Sinyal hasil filter disimpan di: {output_path}")
```
